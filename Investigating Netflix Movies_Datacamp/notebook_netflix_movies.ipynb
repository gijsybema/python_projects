```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

### Question 1. 
What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

```python
# Start coding here! Use as many cells as you like
netflix_df.head()
```

```python
netflix_df.info()
```

```python
#netflix_df.describe()
```

```python
netflix_df['release_year'].dtype
```

```python
netflix_df_90 = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)].copy()
```

```python
netflix_df_90['type'].unique()
```

```python
netflix_df_90_movies = netflix_df_90[netflix_df_90['type'] == 'Movie'].copy()
```

```python
# most frequent duration in the 90s
duration = int(netflix_df_90_movies['duration'].mode().iloc[0])
print(type(duration))
print(duration)
```

### Question 2
A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
netflix_df_90_movies['genre'].unique()
```

```python
netflix_df_90_movies['genre'].value_counts()
```

```python
df_action = netflix_df_90_movies[netflix_df_90_movies['genre'] == 'Action'].copy()
```

```python
short_movie_count = df_action[df_action['duration'] < 90].shape[0]
```
